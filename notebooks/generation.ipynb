### LLM response generation

```python
from huggingface_hub import InferenceClient
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path="../.env") 
hf_token = os.getenv("HF_API_TOKEN")

client = InferenceClient(token=hf_token)
model_name = "deepseek-ai/DeepSeek-R1"

# retrieved context
retrieved_chunks = [
    {"text": "auth_service.py: login_user handles authentication by verifying credentials."},
    {"text": "jwt_utils.py: validate_token checks JWT validity and expiration."}
]
user_query = "How does authentication work?"

context = "\n".join(c["text"] for c in retrieved_chunks)
messages = [
    {"role": "system", "content": "You are a knowledgeable coding assistant."},
    {"role": "user", "content": f"Context:\n{context}\n\nQuestion:\n{user_query}"}
]

response = client.chat_completion(
    model=model_name,
    messages=messages,
    max_tokens=300
)

answer = response.choices[0].message["content"]
print(answer)
```
